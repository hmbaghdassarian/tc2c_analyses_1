```python
import tensorly as tl
import numpy as np
import pandas as pd
```

```python
lr_tensors = list()
for j in range(10):
    lr_matrices = list()
    for i in range(300):
        lr_matrix = np.random.random((12, 12))
        if j in (2,3, 4):
            lr_matrix[:,1] = float('nan')
            lr_matrix[1,:] = float('nan')
        if j in (5,6):
            lr_matrix[:,2] = float('nan')
            lr_matrix[2,:] = float('nan')
        lr_matrices.append(lr_matrix)
    lr_tensor = np.dstack(lr_matrices)
    lr_tensors.append(lr_tensor)

tensor = tl.tensor(np.dstack(lr_tensors).reshape(12,12,300,10))
```

```python
# sparsity
np.count_nonzero(np.isnan(tensor))*100/np.prod(tensor.shape)
```

```python
lr_tensors = list()
for j in range(4):
    lr_matrices = list()
    for i in range(6):
        lr_matrix = np.random.random((5,5))
        if j in (2,3):
            lr_matrix[:,1] = float('nan')
            lr_matrix[1,:] = float('nan')
        lr_matrices.append(lr_matrix)
    lr_tensor = np.dstack(lr_matrices)
    lr_tensors.append(lr_tensor)

tensor = tl.tensor(np.dstack(lr_tensors).reshape(5,5,6,4)) # cell, cell, lr, context
```
